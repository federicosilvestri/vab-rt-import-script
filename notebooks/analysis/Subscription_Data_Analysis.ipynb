```python
tessere_multiple = (
    df.groupby('num_tessera')['nome_socio']
      .nunique()
      .gt(1)
)

tessere_problematiche = tessere_multiple[tessere_multiple].index
tessere_problematiche = tessere_problematiche.map(lambda x: None if not x or len(str(x)) < 2 else x)
tessere_problematiche = tessere_problematiche.dropna()
```

```python
df[df['num_tessera'].isin(tessere_problematiche)][["num_tessera", "nome_socio", "cognome_socio"]].sort_values(['num_tessera', 'nome_socio', 'cognome_socio']).to_csv("duplicati.csv")
```
